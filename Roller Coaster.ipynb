2. Write a function that will plot the ranking of a given roller coaster over time as a line. Your function should take a roller coaster's name and a ranking DataFrame as arguments. Make sure to include informative labels that describe your visualization.

   Call your function with `"El Toro"` as the roller coaster name and the wood ranking DataFrame. What issue do you notice? Update your function with an additional argument to alleviate the problem, and retest your function.

```python
# 2
# Create a function to plot rankings over time for 1 roller coaster
def ranking(name,park,dataframe):
    df=dataframe[(dataframe.Name==name)&(dataframe.Park==park)]
    fig=plt.figure(figsize=(18,8))
    ax=plt.subplot()
    plt.plot(df["Year of Rank"],df["Rank"])
    ax.set_xticks(df["Year of Rank"])
    ax.set_yticks(df["Rank"])
    ax.invert_yaxis()
    plt.title(f"The ranking of  {name} over time")
    plt.xlabel("Year")
    plt.ylabel("Ranking")

    plt.show()
    plt.clf()

# Create a plot of El Toro ranking over time
ranking("El Toro","Six Flags Great Adventure",wood_ranking)
```

```python
def p(number,dataframe):
    lst=[]
    df=dataframe[dataframe.Rank<=number]
    fig=plt.figure(figsize=(18,8))
    for i in range(number):
        new_df=df[df.Name==df.Name[i]]
        ax=plt.subplot()
        plt.plot(new_df["Year of Rank"],new_df["Rank"])
        #ax.set_xticks(range(len(new_df["Year of Rank"])))
        ax.set_yticks(range(1,number+1))
        ax.invert_yaxis()
    
        lst.append(new_df.Name[i])
        
    plt.title(f"Ranking of top {number} Roller Coaster over time")
    plt.xlabel("Year")
    plt.ylabel("Ranking")
    plt.legend(lst)
    plt.show()
    plt.clf()
p(5,wood_ranking)
```

```python
# 6
# Create a function to plot histogram of column values
def histogram(dataframe,col):
    #fig=plt.figure()
    #ax=plt.subplot
    dataframe.hist(col,edgecolor="black")
    plt.title(f"The {col} of roller coaster")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
    plt.clf()



# Create histogram of roller coaster speed
histogram(roller_costers,"speed")
# Create histogram of roller coaster length
histogram(roller_costers,"length")
# Create histogram of roller coaster number of inversions
histogram(roller_costers,"num_inversions")
# Create a function to plot histogram of height values
def height_coaster(dataframe):
    #height=dataframe["height"].dropna()
    df=dataframe[dataframe.height<=140].dropna()
    
    df.hist("height",edgecolor="black")
    plt.title("The height of roller coaster")
    plt.xlabel("Height")
    plt.ylabel("Count")
    plt.show()
    plt.clf()
# Create a histogram of roller coaster height
height_coaster(roller_costers)
```

```python
# 7
# Create a function to plot inversions by coaster at park
def coster_names(park_name,dataframe):
    df=dataframe[dataframe.park==park_name]
    trimed_df=df[["name","num_inversions"]]
    df=dataframe[dataframe==park_name]
    trimed_df=trimed_df.sort_values("num_inversions",ascending=False)
 
    fig=plt.figure(figsize=(18,8))
    ax=plt.subplot()
    plt.bar("name","num_inversions",data=trimed_df)
    ax=plt.subplot()
    ax.set_xticklabels(trimed_df["name"],rotation=90)
    plt.title("Number of inversions of {park_name}.")
    plt.xlabel("Name of Roller Coaster")
    plt.ylabel("Count")
    plt.show()
    plt.clf()


# Create barplot of inversions by roller coasters
coster_names('Six Flags Great Adventure',roller_costers)

```

9. `.scatter()` is another useful function in matplotlib that you might not have seen before. `.scatter()` produces a scatter plot, which is similar to `.plot()` in that it plots points on a figure. `.scatter()`, however, does not connect the points with a line. This allows you to analyze the relationship between two variables. Find [`.scatter()`'s documentation here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html).

```python
def speciality_company(dataframe,manufacturar_name):
    df=dataframe[dataframe.manufacturer==manufacturar_name]
    hight_varience=df.height.nunique()
    length_varience=df.length.nunique()
    speed_varience=df.speed.nunique()
    seatingType_varience=df.seating_type.nunique()
    inversion_varience=df.num_inversions.nunique()
    lst=[hight_varience,length_varience,speed_varience,seatingType_varience,inversion_varience]
    lab=["heights","lengths","speed","seating_type","inversions"]
    plt.bar(lab,lst)
    plt.title(f"The varience of different structures of roller coster of {manufacturar_name} company focused on")
    plt.xlabel("Different configaretion of Roller Coaster")
    plt.ylabel("Varience Count")
    plt.show()
    plt.clf()
speciality_company(roller_costers,"Vekoma")


def speciality_park(dataframe,park_name):
    df=dataframe[dataframe.park==park_name]
    hight_varience=df.height.nunique()
    length_varience=df.length.nunique()
    speed_varience=df.speed.nunique()
    seatingType_varience=df.seating_type.nunique()
    inversion_varience=df.num_inversions.nunique()
    lst=[hight_varience,length_varience,speed_varience,seatingType_varience,inversion_varience]
    lab=["heights","lengths","speed","seating_type","inversions"]
    plt.bar(lab,lst)
    plt.title(f"The varience of different structures of roller coster of {park_name} park focused on")
    plt.xlabel("Different configaretion of Roller Coaster")
    plt.ylabel("Varience Count")
    plt.show()
    plt.clf()
speciality_park(roller_costers,"Parc Asterix")
```
